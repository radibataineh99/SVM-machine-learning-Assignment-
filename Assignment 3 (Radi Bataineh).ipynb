# Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score,recall_score,precision_score
import seaborn as sn
from sklearn.model_selection import GridSearchCV
```

# Importing the dataset

```python
dataset = pd.read_csv('SVMdataset.csv')
x = dataset.iloc[:, :2].values
y = dataset.iloc[:, 2].values
```

# Data Set Visualization

```python
from matplotlib.colors import ListedColormap
X_set, y_set = x, y
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min()-0.2, stop = X_set[:, 0].max()+0.2, step = 0.25),
                     np.arange(start = X_set[:, 1].min()-0.1, stop = X_set[:, 1].max()+0.1, step = 0.25))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
k=["-ve","+ve"]
ma=['o','x']
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'blue'))(i), marker=ma[i],label = k[j])
plt.title('SVM(Data Set)')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```

##  plots showing the decision boundary (training)

```python
X = x_train 
y = y_train

def make_meshgrid(x, y, h=.02):
    x_min, x_max = x.min()-0.4 , x.max()+0.4 
    y_min, y_max = y.min()-0.4 , y.max()+0.4 
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    return xx, yy

def plot_contours(ax, clf, xx, yy, **params):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out

fig, ax = plt.subplots()
# title for the plots
title = ('SVM(traing Set)')
# Set-up grid for plotting.
X0, X1 = X[:, 0], X[:, 1]
xx, yy = make_meshgrid(X0, X1)

plot_contours(ax, linearSVM, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)
ax.scatter(X0, X1,y==1, c='blue',marker='x', cmap=plt.cm.coolwarm,linewidth=8, edgecolors='k',label="+ve")
ax.scatter(X0, X1,y==0, c='red',marker='o', cmap=plt.cm.coolwarm,linewidth=8 , edgecolors='k',label="-ve")
ax.set_ylabel('X2')
ax.set_xlabel('X1')
ax.set_xticks(())
ax.set_yticks(())
ax.set_title(title)
ax.legend()
plt.show()
```
